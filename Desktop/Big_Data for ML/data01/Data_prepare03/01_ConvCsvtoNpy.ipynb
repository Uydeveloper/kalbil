```python
import pandas as pd
import numpy as np
```

```python
npyname = "kddi_all_new_0320max" #kddi_00_new_0315_15min
datedir = "230316"
```

```python
# 交通量データの読込み
csv_file = r'./'+datedir+'/'+npyname+'.csv'
df = pd.read_csv(csv_file, dtype={'date':str}, index_col=0, parse_dates=True)
```

```python
# 読み込んだデータの確認
df
```

```python
# 15分単位でGroup byして、結果をagg関数にsum関数を渡して集計
totalling_df = df.groupby([pd.Grouper(level='times', freq='30T'), 'osm_id']).agg(np.sum) #gridcode
```

```python
# 集計したデータの確認
totalling_df
```

```python
totalling_df.to_csv(datedir+'/'+npyname+'totalling_df_osm.csv')
```

```python
# 行列の入れ替え
output_df = totalling_df.unstack(fill_value=0)
```

```python
output_df = output_df.fillna(0)
```

```python
# 行列入れ替え後のデータを確認
output_df
```

```python
output_df.to_csv(datedir+'/'+npyname + "_header.csv")
```

```python
# numpy_arrayに変換
output_nd = output_df.to_numpy()
```

```python
output_nd.astype(np.double)
```

```python
# numpy_arrayに変換したデータの確認
output_nd
```

```python
#正規化データの保存
from sklearn.preprocessing import MinMaxScaler
# データの正規化
# データを0～1の範囲に正規化）
scaler = MinMaxScaler(feature_range=(0, 1))
np.save(datedir+'/'+npyname + '_01weight.npy', scaler.fit_transform(output_nd))
```

```python
# データの保存
np.save(datedir+'/'+npyname + '_01param.npy', output_nd)
```
